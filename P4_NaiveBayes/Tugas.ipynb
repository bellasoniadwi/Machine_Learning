```python
import numpy as np
import pandas as pd

#load data ke Csv
df = pd.read_csv("/content/spam.csv", encoding='latin-1')

df.head()
```

```python
#Drop 3 kolom terakhir dengan fungsi iloc
df = df.drop(df.iloc[:, 2:], axis=1)

#Cek data
df.head()
```

```python
#Data untuk rename kolom v1 dan v2
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

#Rename nama kolom v1 dan v2
df = df.rename(columns=new_cols)

#Cek data
df.head()
```

```python
#Cek jumlah Data Per Kelas
print(df['Labels'].value_counts())
print('\n')

#Cek Kelengkapan Data
print(df.info())
print('\n')

#Cek Statistik Deskriptif
print(df.describe())
```

```python
#Data untuk label
new_labels = {
    'spam': 1,
    'ham': 0
}

#Encode label
df['Labels'] = df['Labels'].map(new_labels)

#Cek data
df.head()
```

```python
x = df['SMS'].values
y = df['Labels'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

#Split data training dan testing
x_train_cv, x_test_cv, y_train_cv, y_test_cv = train_test_split(x, y, test_size=0.2, random_state=50)

#menginisiasi CountVectorizer dengan stop_words
bow_1 = CountVectorizer(stop_words='english')

#fitting dan transform x_train dengan CountVectorizer
x_train_cv = bow_1.fit_transform(x_train_cv)

#transform x_test
x_test_cv = bow_1.transform(x_test_cv)

#display x_train setelah pembobotan CountVectorizer
print(x_train_cv)
```

```python
#menampilkan kumpulan token yang sudah di eliminasi stop_words
token_1 = bow_1.get_feature_names_out()
print(token_1)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinominalNB
mnb = MultinomialNB()

# Fit model
mnb.fit(x_train_cv, y_train_cv)

# Prediksi dengan data training
y_pred_train_cv = mnb.predict(x_train_cv)

# Evaluasi akurasi data training
acc_train_cv = accuracy_score(y_train_cv, y_pred_train_cv)

# Prediksi dengan data testing
y_pred_test_cv = mnb.predict(x_test_cv)

# Evaluasi akurasi data testing
acc_test_cv = accuracy_score(y_test_cv, y_pred_test_cv)

# Print akurasi
print(f'Hasil akurasi data train: {acc_train_cv}')
print(f'Hasil akurasi data test: {acc_test_cv}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

#Split data training dan testing
x_train_tiv, x_test_tiv, y_train_tiv, y_test_tiv = train_test_split(x, y, test_size=0.2, random_state=50)

#menginisiasi TfidfVectorizer dengan stop_words
bow_2 = TfidfVectorizer(stop_words='english')

#fitting dan transform x_train dengan CountVectorizer
x_train_tiv = bow_2.fit_transform(x_train_tiv)

#transform x_test
x_test_tiv = bow_2.transform(x_test_tiv)

#display x_train setelah pembobotan CountVectorizer
print(x_train_tiv)
```

```python
# Menampilkan kumpulan token yang sudah di eliminasi stop_words
token_2 = bow_2.get_feature_names_out()
print(token_2)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinominalNB
mnb = MultinomialNB()

# Fit model
mnb.fit(x_train_tiv, y_train_tiv)

# Prediksi dengan data training
y_pred_train_tiv = mnb.predict(x_train_tiv)

# Evaluasi akurasi data training
acc_train_tiv = accuracy_score(y_train_tiv, y_pred_train_tiv)

# Prediksi dengan data testing
y_pred_test_tiv = mnb.predict(x_test_tiv)

# Evaluasi akurasi data testing
acc_test_tiv = accuracy_score(y_test_tiv, y_pred_test_tiv)

# Print akurasi
print(f'Hasil akurasi data train: {acc_train_tiv}')
print(f'Hasil akurasi data test: {acc_test_tiv}')
```

```python
# Print Perbandingan Hasil Akurasi Data Train dan Data Test dengan CountVectorizer vs TfidfVectorizer
print('COUNT VECTORIZER')
print(f'Hasil akurasi data train : {acc_train_cv}')
print(f'Hasil akurasi data test: {acc_test_cv}\n')

print('TF-IDF VECTORIZER')
print(f'Hasil akurasi data train : {acc_train_tiv}')
print(f'Hasil akurasi data test: {acc_test_tiv}\n')
```
